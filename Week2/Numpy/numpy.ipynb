```python
#  1 dimensional array
a=np.array([1,2,3,4,5])
print(a[0])
print(a[-1])
#  2 dimensional array
b=np.array([[2,3,4,9],[4,3,5,6]])
print(b[0,0])
print(b[-1,-1])
#  2 dimensional array
c=a=np.array([[[2,3,4,9],[4,3,5,6]],[[4,6,4,3],[4,7,8,3]]])
print(c[0,1,2])
print(c[-1,-2,-3])
```

```python
a=np.array([[1,2,3,4,1],[4,5,6,7,2]],dtype=np.int64)
b=np.array([1,2,3,4,5,6,7,8])
print(a.ndim)
print(a.shape)
print(a.size)
print(a.itemsize)
print(a.data)
print(a.dtype)
print(a.strides)
print(a.reshape(2,5))
print(np.resize(a,(5,2)))
print(np.reshape(b,(2,2,2)))
```

```python
#Joins in array
#1.concatenate()
#2.hstack()
#3.vstack()
```

```python
#concatenate()
a=np.array([2,3,4,9])
b=np.array([4,3,5,6])
c=np.concatenate([a,b])
print(c)
a=np.array([[2,3,4,9],[4,3,5,6]])
b=np.array([[4,6,4,3],[4,7,8,3]])
c=np.concatenate([a,b])
c
```

```python
#axis=0,column wise
a=np.array([[2,3,4,9],[4,3,5,6]])
b=np.array([[4,6,4,3],[4,7,8,3]])
c=np.concatenate([a,b],axis=0)
print(c)
#axis=1,row wise
a=np.array([[2,3,4,9],[4,3,5,6]])
b=np.array([[4,6,4,3],[4,7,8,3]])
c=np.concatenate([a,b],axis=1)
print(c)
```

```python
#vstack()
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.vstack((a,b))
print(c)
#hstack()
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.hstack([a,b])
print(c)
```

```python
# working with matrix
a=np.array([[1,2,3],[4,5,6],[6,7,8]])
b=np.array([[9,8,7],[6,5,4],[3,2,1]])
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
```

```python
a=np.array([[1,2,3],[4,5,6],[6,7,8]])
b=np.array([[9,8,7],[6,5,4],[3,2,1]])
print(np.r_[(a,b)])   #row_wise
print(np.c_[(a,b)])   #column_wise
```

```python
# it converts a multi‑dimensional array into a 1‑D array.
a=np.array([[1,2,3],[4,5,6],[6,7,8]])
b=np.array([[[9,8,7],[6,5,4],[3,2,1]]])
print(np.reshape(a,-1))
print(b.flatten())
```

```python
#Expanding Array
array=np.array([[2,5,3],[8,3,9],[1,4,7]])
print(np.expand_dims(array,axis=(0,1)))
```

```python
# squeeszing
a=np.array([[[1],[2],[3]]])
print(a)
print(a.shape)
b=np.squeeze(a,axis=0)
print(a)
print(a.shape)
```
